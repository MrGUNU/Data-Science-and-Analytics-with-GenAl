```python
# Question 5 (Left rotation)
a = [10,20,30,40,50]

for i in range(len(a)-1):
    a[i],a[i+1] = a[i+1],a[i]

print(a)
```

```python
# (Right rotation)
a = [10,20,30,40,50]

for i in range(len(a)-1,0,-1):
    a[i],a[i-1] = a[i-1],a[i]

print(a)
```

```python
# Question 6 (k times rotation)
k = int(input("How many times you rotate: "))
a = [10,20,30,40,50]

for i in range(k):
    for i in range(len(a)-1):
        a[i],a[i+1] = a[i+1],a[i]

print(a)
```

```python
# Question 7 (Reverse the list)
a = [10,20,30,40,50]
b = len(a)-1

for i in range(len(a)//2):
    a[i],a[b] = a[b],a[i]
    b = b-1

print(a)
```

```python
# Liner search
a = [23,4,12,45,76,98,33,6,50]

for i in range(len(a)):
    if a[i] == 6:
        print(f"Found in index {i}")
        break
else:
    print("Not found")
```

```python
# Bubble sort
a = [23,43,75,2,10,18,47,90,89,100,150,15, 20,30]

for j in range(len(a)):
    for i in range(len(a)-1-j):
        if a[i] > a[i+1]:
            a[i],a[i+1] = a[i+1],a[i]
print(a)
```

```python
# Selection sort
a = [23,43,75,2,10,18,47,90,89,100,150,15, 20,30]

for i in range(len(a)-1):
    j = i+1
    min = i
    for k in range(j,len(a)):
        if a[k] < a[min]:
            min = k
    a[i],a[min] = a[min],a[i]
print(a)
```
