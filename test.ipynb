```python
# 1st way to solve prime number check

n = int(input("Enter the number: "))

count = 0

for i in range(1, n+1):
    if n % i == 0:
        count = count + 1

if count == 1:
    print(f"{n} is a unity number")
elif count == 2:
    print(f"{n} is a prime number")
else:
    print(f"{n} is a composite number")    
```

```python
# 2nd way to solve prime number check

n = int(input("Enter the number: "))

for i in range(2,n):
    if n % i == 0:
        print(f"{n} is a composite number")
        break
    
else:
    print(f"{n} is a prime number")
```

```python
n = int(input())
n = abs(n)

even = 0
odd = 0

# Special case for 0
if n == 0:
    even = 1
else:
    while n > 0:
        digit = n % 10

        if digit % 2 == 0:
            even += 1
        else:
            odd += 1
        
        n = n // 10

print("Even:", even)
print("Odd:", odd)
```

```python
# Check 2 words are Anagrams or Not Anagrams

s1 = input().strip()
s2 = input().strip()

if len(s1) != len(s2):
    print("Not Anagrams")
else:
    freq = {}

    for ch in s1:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1

    for ch in s2:
        if ch in freq:
            freq[ch] -= 1
        else:
            print("Not Anagram")
            break

    else:
        for val in freq.values():
            if val != 0:
                print("Not Anagram")
                break
        else:
            print("Anagram")
```

```python
# All perfect square within a given range

start, end = map(int, input().split())

for num in range(start, end + 1):
    root = int(num ** 0.5)
    if root * root == num:
        print(num, end=" ")
```

```python
n = int(input())
c = n
sum_fact = 0

while n > 0:
    digit = n % 10

    fact = 1
    i = 1
    while i <= digit:
        fact *= i
        i += 1
    sum_fact += fact
    n //= 10
if sum_fact == c:
    print("Strong Number")
else:
    print("Not a Strong Number")
```
